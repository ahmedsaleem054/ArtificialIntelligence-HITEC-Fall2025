```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers."
    elif n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)



num = int(input("Enter a non-negative integer: "))

result = factorial(num)

print(f"The factorial of {num} is {result}")

```

```python
def get_unique_elements(input_list):
    unique_list = []
    for item in input_list:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list

sample_list = [1, 2, 3, 3, 4, 4, 5, 3, 7, 5]

unique_list = get_unique_elements(sample_list)

print("Unique List:", unique_list)
```

```python
def get_even_numbers(n):
    even_numbers = []
    for i in range(1, n + 1):
        even_numbers.append(i * 2)
    print("First", n, "even numbers:", even_numbers)
    return even_numbers

n = int(input("Enter how many even numbers you want: "))

evens = get_even_numbers(n)

even_sum = sum(evens)

product = 1
for num in evens:
    product *= num

print("Sum of even numbers:", even_sum)
print("Product of even numbers:", product)
```

```python
def match_strings(str1, str2):
    """
    Check if two strings match exactly (case-sensitive).
    """
    if str1 == str2:
        return True
    else:
        return False

string1 = input("Enter the first string: ")
string2 = input("Enter the second string: ")

if match_strings(string1, string2):
    print("✅ The strings match exactly.")
else:
    print("❌ The strings do not match.")
```

```python
def login_system(username, password):

    stored_username = "admin"
    stored_password = "12345"

    if username == stored_username and password == stored_password:
        return True
    else:
        return False

user = input("Enter your username: ")
pwd = input("Enter your password: ")

if login_system(user, pwd):
    print("✅ Login successful! Welcome,", user)
else:
    print("❌ Login failed! Invalid username or password.")
```
